```python
import pandas as pd
import re
import string

df = pd.read_csv("raw_asr_vs_reference1.csv")

# --------------------------
# Hindi Number Dictionary
# --------------------------
hindi_numbers = {
    "एक": 1, "दो": 2, "तीन": 3, "चार": 4, "पांच": 5,
    "छह": 6, "सात": 7, "आठ": 8, "नौ": 9, "दस": 10,
    "ग्यारह": 11, "बारह": 12, "तेरह": 13, "चौदह": 14,
    "पंद्रह": 15, "सोलह": 16, "सत्रह": 17, "अठारह": 18,
    "उन्नीस": 19, "बीस": 20, "तीस": 30, "चालीस": 40,
    "पचास": 50, "साठ": 60, "सत्तर": 70, "अस्सी": 80,
    "नब्बे": 90, "सौ": 100, "हजार": 1000
}

# --------------------------
# Number Normalization
# --------------------------
def normalize_numbers(text):
    words = text.split()
    new_words = []

    for w in words:
        if w in hindi_numbers:
            new_words.append(str(hindi_numbers[w]))
        else:
            new_words.append(w)

    return " ".join(new_words)

# --------------------------
# Remove repetitions
# --------------------------
def remove_repetition(text):
    return re.sub(r'\b(\w+)( \1\b)+', r'\1', text)

# --------------------------
# Remove fillers
# --------------------------
fillers = ["uh", "um", "ah", "hmm", "yeah", "okay", "ok"]

def remove_fillers(text):
    words = text.split()
    words = [w for w in words if w.lower() not in fillers]
    return " ".join(words)

# --------------------------
# Remove punctuation
# --------------------------
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

# --------------------------
# Full Cleaning Pipeline
# --------------------------
def clean_text(text):
    text = str(text)
    text = text.lower()

    text = remove_punctuation(text)
    text = remove_repetition(text)
    text = normalize_numbers(text)
    text = remove_fillers(text)

    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    return text

# Apply cleaning
df["cleaned_asr"] = df["raw_asr"].apply(clean_text)
df["cleaned_ref"] = df["reference"].apply(clean_text)

# Save
df.to_csv("cleaned_asr5.csv", index=False)

print("Cleaned file saved")
```
